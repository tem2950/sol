```python
import numpy as np
import matplotlib.pyplot as plt

#parametres
N=1000  #total population
alpha=20 #infection rate
a=alpha/N #            
g=0.02 #recovery rate

#initial conditions\
S0=990 ;I0=10

dt=0.1
tmax=100
nsteps=int((tmax/dt)+1)
S=np.zeros(nsteps) #             kai#    .   #         ,#       tmax   # dt
I=np.zeros(nsteps)

S[0]=S0
I[0]=I0

S_rate = np.zeros(int(tmax/dt)+1)
I_rate = np.zeros(int(tmax/dt)+1)

S_rate[0] = 0
I_rate[0] = 0

for i in range(1,nsteps):
    S[i]=S[i-1]-a*S[i-1]*I[i-1]*dt+g*I[i-1]*dt #    dt                   euler
    I[i]=I[i-1]+a*S[i-1]*I[i-1]*dt-g*I[i-1]*dt

for i in range(1,nsteps):
    S_rate[i] = -a*S[i-1]*I[i-1] + g*I[i-1] 
    I_rate[i] = a*S[i-1]*I[i-1] - g*I[i-1] 

fig, axs = plt.subplots(4, 1, figsize=(8, 12))

axs[0].set_title('a={0:},g={1:}'.format(a,g))
axs[0].plot(S[:100],'g',label='Susceptible')
axs[0].plot(I[:100],'r',label='Infected')
axs[0].set_xlabel('time')
axs[0].set_ylabel('S,I')
fig.tight_layout()
axs[0].legend()

axs[1].set_title('rate of S')
axs[1].plot(S,S_rate,'g',label='rate of S')
axs[1].set_xlabel('susceptible')
axs[1].set_ylabel('dS/dt')
fig.tight_layout()
axs[1].legend()

axs[2].set_title('rate of I')
axs[2].plot(I,I_rate,'r',label='rate of I')
axs[2].set_xlabel('infected')
axs[2].set_ylabel('dI/dt')
fig.tight_layout()
axs[2].legend()

axs[3].set_title('S-V')
axs[3].scatter(S[:-1], I[:-1], color='purple',s=3)
axs[3].set_xlabel('Susceptible')
axs[3].set_ylabel('Infected')
axs[3].set_title('Final State of SIS Model')




plt.show()
```
